# Now lets merge the SEM dataframe with the Raman data

## First import the Raman data

```python
Raman_df=pd.read_excel(compilation_folder+slash+'Ulu_Raman_averaged.xlsx')
Raman_df = Raman_df.loc[:, ~Raman_df.columns.str.contains('Unnamed')]

Raman_df
```

## Import the olivine data

```python
OlivineEDS=pd.read_excel(compilation_folder+slash+'EDS_olivine.xlsx')
OlivineEDS =OlivineEDS.loc[:, ~OlivineEDS.columns.str.contains('Unnamed')]
# OlivineEDS=OlivineEDS[~OlivineEDS["Sample Name"].str.contains("K20")]
# OlivineEDS.to_clipboard(excel=True,index=False)
OlivineEDS.head()
```

```python
Raman_df['Name_on_SEM']
```

## Merge the olivines to the FI

```python
sorted_df_renamed = OlivineEDS.rename(columns={'Sample Name': 'Name_on_SEM'})

sorted_df_renamed = sorted_df_renamed.rename(columns=lambda x: x + '_Ol' if x != 'Name_on_SEM' else x)

combined_df = Raman_df.merge(sorted_df_renamed, on='Name_on_SEM', how='left')
combined_df
```

## Now let's merge the glass data for Ulu

### Now let's get the Ulu data (1974) from Wieser et al. 2021

```python
Ulu_all=pd.read_excel(MasterFolder+slash+"Data"+slash+"Data_Ulu_mounts"+slash+"41467_2019_13635_MOESM4_ESM_mod.xlsx",
                         sheet_name='Glass Chemistry').dropna(axis=0,how='all')
Ulu_all['Eruption']=Ulu_all['Eruption'].astype('str')

Ulu_all_filt=Ulu_all.loc[:,(~Ulu_all.columns.str.contains('ppm|corrected|Duration|Spot|Batch'))]
Ulu_all_filt.columns = Ulu_all_filt.columns.str.replace('_Liq', '')
Ulu_all_filt=Ulu_all_filt.rename(columns={'Eruption':'Sample','Glass Fragment ID':'Sample Name','Total':'Total_wt%','Mg Number':'Mg#'})
Ulu_all_filt['Sample'] = Ulu_all_filt['Sample'].astype(str).str.split('.').str[0]
Ulu_all_filt_renamed =Ulu_all_filt.rename(columns=lambda x: x + '_Gl' if x != 'Sample' else 'Sample')

Ulu_all_filt_renamed.to_excel(compilation_folder+'/'+"UluGlass_averaged.xlsx")


Ulu_all_filt_renamed
```

### And let's get the Sides et al. (2014)

```python
Sides_glasses_og=pd.read_excel(MasterFolder +slash+"Data_processing_notebooks"+slash+'Data_processing_Keana_Uwe_Litt'+ 
                               slash +'Notebook_Compiled_data'+slash+"UW_L84_K21_averaged.xlsx").dropna(axis=0,how='all')

Sides_glasses_og =Sides_glasses_og.loc[:, ~Sides_glasses_og.columns.str.contains('Unnamed')].copy()
Sides_glasses_og
```

### Combine the glasses together

```python
merged_glasses=pd.concat([Ulu_all_filt_renamed,Sides_glasses_og],axis=0)
merged_glasses
```

## Now merge and export the merged data

```python
combined_df['Sample'] = combined_df['Sample'].astype(str)
combined_df_wglass = combined_df.merge(merged_glasses, on='Sample', how='left')
# combined_df_wglass.to_clipboard(excel=True)

combined_df_wglass.to_excel(compilation_folder+slash+'Ulu_Raman_EDS_Gl_Merged.xlsx')

combined_df_wglass
```

# Now let's merge the EBSD

## First load the EBSD data from Wieser et al and merge the GOS params with Slip System params

```python
## This loads data from Wieser et al and merges GOS with slip systems and linear intercept data.

ebsd_all_kil=pd.read_excel(MasterFolder+slash+"Data"+slash+"Data_Ulu_mounts"+slash+"EBSD"+slash+"PWCambridge"+slash+"Merged_GOS_Parameters_edited_Kil.xlsx", 
                           sheet_name="Sheet1")
ebsd_all_kil['length_norm']=ebsd_all_kil['subBoundaryLength (um)']/ebsd_all_kil['equivalentRadius (um)']
ebsd_all_kil['to_merge']= ebsd_all_kil['Filename'] + '_' + np.round(ebsd_all_kil['GOS'], 3).astype(str)

slip_sys_kil=pd.read_excel(MasterFolder+slash+"Data"+slash+"Data_Ulu_mounts"+slash+"EBSD"+slash+"PWCambridge"+slash+"Merged_GOS_Parameters_edited_Kil.xlsx",
                            sheet_name='SlipSystemID')

slip_sys_kil['file_mod']=slip_sys_kil['file'].str.replace('\'', '')
slip_sys_kil['to_merge']= slip_sys_kil['file_mod'] + '_' + np.round(slip_sys_kil['GOS'], 3).astype(str)

# Average out the linear intercepts
slip_sys_kil['mean_length'] = slip_sys_kil.loc[:, 'len1':'len20'].mean(axis=1, skipna=True)
slip_sys_kil['median_length'] = slip_sys_kil.loc[:, 'len1':'len20'].median(axis=1, skipna=True)

## combine
combined_df_kil=pd.merge(ebsd_all_kil, slip_sys_kil, on='to_merge', how='outer', suffixes=('', '_dup'))
combined_df_kil = combined_df_kil.loc[:, ~combined_df_kil.columns.str.endswith('_dup')]

combined_df_kil=combined_df_kil.fillna(0)
combined_df_kil
```

## We also have to load an merge the new EBSD data from  UCB

```python
ebsd_ucb=pd.read_excel(MasterFolder+slash+"Data"+slash+"Data_Ulu_mounts"+slash+"EBSD"+slash+"UCB"+slash+"EBSD_combined_UCB_all.xlsx", 
                           sheet_name="Sheet1")
cam_ucb_ebsd=pd.concat([combined_df_kil, ebsd_ucb])

cam_ucb_ebsd.to_excel(MasterFolder+slash+"Data"+slash+"Data_Ulu_mounts"+slash+"EBSD"+slash+"Combined_EBSD"+slash+'Ulu_EBSD_combined_all.xlsx')
cam_ucb_ebsd
```

## Now let's merge with the rest of the data from above

```python
EBSD_data=pd.read_excel(MasterFolder +slash+"Data"+slash+"Data_Ulu_mounts"+slash+"EBSD"+slash+"Combined_EBSD"+slash+'Ulu_EBSD_combined_all.xlsx')
EBSD_data=EBSD_data.rename(columns={'EBSD_name':'EBSD_simple_name','grainID':'EBSD_grainID','Filename':'EBSD_filename'})
EBSD_data['EBSD_simple_name']=EBSD_data['EBSD_filename'].copy()

final_df=pd.merge(left=combined_df_wglass,on=['EBSD_simple_name','EBSD_grainID'],right=EBSD_data,suffixes=('', '_EBSDdata'),how='left')

# ##### Export fully merge data ####

final_df.to_excel(compilation_folder+slash+"Ulu_Final_data_wEBSD.xlsx")
final_df
```
