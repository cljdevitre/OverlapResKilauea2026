# This compiles all the Raman data into one file, and exports

## First create the function

```python


## This function compiles raman sessions found in the master folder (even in subdirectories of the specified folder)

def compile_Raman_sessions(*, path=None, sheet_name='Sheet1', keyword='Raman_session'):
    df = pd.DataFrame()

    for root, dirs, files in os.walk(path):
        for file in files:
            if keyword in file and file.endswith('.xlsx'):
                print(file)
                file_path = os.path.join(root, file)
                sheets = pd.read_excel(file_path, sheet_name)
                df = pd.concat([df, sheets], axis=0, ignore_index=True)
                unnamed_columns = [col for col in df.columns if 'Unnamed' in col]
                df = df.drop(columns=unnamed_columns)

    return df
```

## Now run the function

```python
Ulu = compile_Raman_sessions(path=MasterFolder+slash+"Data"+slash+"Data_Ulu_mounts"+slash+"Raman",sheet_name='Sheet1',keyword='Raman_session')
```

```python
rep_FI_names=pd.read_excel('Helper files/'+'Helperfile_Ulu_mounts.xlsx',sheet_name='Sheet1')
Ulu_reps = pd.merge(Ulu, rep_FI_names, on=['filename', 'date_object','sec since midnight'])
Ulu_reps = Ulu_reps[~Ulu_reps['filename'].str.contains('46m')]
Ulu_reps
```

## Now let's take mean of repeated analyses

```python
grouped = Ulu_reps.groupby('FI_name')

# Separate numeric and non-numeric columns
numeric_cols = Ulu_reps.select_dtypes(include='number')
non_numeric_cols = Ulu_reps.select_dtypes(exclude='number')

# Aggregate numeric columns by mean and standard deviation
numeric_Ulu_averaged_mean = grouped[numeric_cols.columns].mean()
numeric_Ulu_averaged_std = grouped[numeric_cols.columns].std()

# Take the first instance for non-numeric columns
non_numeric_Ulu_averaged = grouped[non_numeric_cols.columns].first()

# Concatenate Ulu_averageds and reindex columns to the original order
Ulu_averaged = pd.concat([numeric_Ulu_averaged_mean, non_numeric_Ulu_averaged], axis=1)
Ulu_averaged = Ulu_averaged.reindex(columns=Ulu_reps.columns)

# Add standard deviation columns after reindexing
Ulu_averaged = pd.concat([Ulu_averaged, numeric_Ulu_averaged_std.add_suffix('_Raman_STD')], axis=1)

# Reset index to get a DataFrame
Ulu_averaged = Ulu_averaged.reset_index(drop=True)

row_counts = grouped.size().reset_index(name='row_count')

# Add a new column "averaged?" based on the row count
row_counts['averaged?'] = 'No'
row_counts.loc[row_counts['row_count'] > 1, 'averaged?'] = 'Yes'

# Merge the row counts back to the aggregated DataFrame
Ulu_averaged = pd.merge(Ulu_averaged, row_counts[['FI_name', 'averaged?']], on='FI_name', how='left')

# If there are missing values (groups with a single row), fill them with 'No'
Ulu_averaged['averaged?'] = Ulu_averaged['averaged?'].fillna('No')

#exclude a weird null average spectrum
# Ulu_averaged=Ulu_averaged[~(Ulu_averaged['filename'].str.contains("186-9-118-A"))]
Ulu_averaged.to_excel(compilation_folder+'/'+"Ulu_Raman_averaged.xlsx")

Ulu_averaged
```
