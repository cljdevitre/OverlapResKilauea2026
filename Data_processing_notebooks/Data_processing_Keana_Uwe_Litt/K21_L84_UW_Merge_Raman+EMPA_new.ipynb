### Function to average glass data

```python
### Create reusable function to average glass data and calculate std 
# def combine_strings_with_counts(series):
#     """
#     Combine string values into 'value(count)' joined by underscores.
#     NaNs are represented as ' ' unless all values are NaN.
#     """
#     if series.isna().all():
#         return np.nan

#     filled = series.fillna(" ").astype(str)
#     counts = filled.value_counts()

#     return "_".join(f"{val}({cnt})" for val, cnt in counts.items())



def average_by_col(
    Data,
    group_col="Sample",
    suffix="_Gl",
    averaged_col_name="averaged?",
    non_numeric_behavior="first"  # NEW
):
    """
    Group a DataFrame by `group_col`, average numeric columns,
    handle non-numeric columns according to `non_numeric_behavior`,
    append STD columns, flag whether rows were averaged,
    and rename columns with a suffix.

    non_numeric_behavior:
        - "first" : keep first value per group (default)
        - "last"  : keep last value per group
        - "skip"  : drop non-numeric columns entirely
    """

    grouped = Data.groupby(group_col)

    # Separate numeric and non-numeric columns
    numeric_cols = Data.select_dtypes(include='number')
    non_numeric_cols = Data.select_dtypes(exclude='number')

    # Aggregate numeric columns
    numeric_Data_averaged_mean = grouped[numeric_cols.columns].mean()
    numeric_Data_averaged_std = grouped[numeric_cols.columns].std()

    if non_numeric_behavior == "first":
        non_numeric_Data_averaged = grouped[non_numeric_cols.columns].first()

    elif non_numeric_behavior == "last":
        non_numeric_Data_averaged = grouped[non_numeric_cols.columns].last()

    elif non_numeric_behavior == "skip":
        non_numeric_Data_averaged = (
            grouped[[group_col]].first()
            if group_col in non_numeric_cols.columns
            else None
        )
    else:
        raise ValueError(
            "non_numeric_behavior must be one of: 'first', 'last', 'skip'"
        )

    # Concatenate aggregated data
    Data_averaged = pd.concat(
        [numeric_Data_averaged_mean, non_numeric_Data_averaged]
        if non_numeric_Data_averaged is not None
        else [numeric_Data_averaged_mean],
        axis=1
    )

    if non_numeric_behavior != "skip":
        Data_averaged = Data_averaged.reindex(columns=Data.columns)

    # Add standard deviation columns
    Data_averaged = pd.concat(
        [Data_averaged, numeric_Data_averaged_std.add_suffix('_STD')],
        axis=1
    )

    # Reset index
    Data_averaged = Data_averaged.reset_index(drop=True)

    # Row counts
    row_counts = grouped.size().reset_index(name='row_count')
    row_counts[averaged_col_name] = 'No'
    row_counts['#of analyses (when averaged)'] = pd.NA
    row_counts.loc[row_counts['row_count'] > 1, '#of analyses (when averaged)'] = row_counts['row_count']
    row_counts.loc[row_counts['row_count'] > 1, averaged_col_name] = 'Yes'

    # Merge averaged flag
    Data_averaged = pd.merge(
        Data_averaged,
        row_counts[[group_col, averaged_col_name,'#of analyses (when averaged)']],
        on=group_col,
        how='left'
    )

    Data_averaged[averaged_col_name] = Data_averaged[averaged_col_name].fillna('No')
    Data_averaged['non_numeric_column_behavior_used (First/Last instance kept or skipped)'] = non_numeric_behavior
    # Rename columns
    Data_averaged_renamed = Data_averaged.rename(
        columns=lambda x: x + suffix if x != group_col else group_col
    )

    return Data_averaged_renamed
```

### Now import the glass data for each eruption

### First, import the glass data for Kulanaokuaiki (unpublished Wieser data)

```python
glasses=pd.read_excel(MasterFolder+slash+'Data'+slash+'Data_Kea_Uwe_Litt'+slash+'Data_all_EPMA'+
                          slash+'Wieser_Glasses_Explosive_Combined.xlsx').dropna(axis=0,how='all')
glasses['Eruption']='Kulanaokuaiki (400-950 CE)'

glasses.columns = glasses.columns.str.replace('_Liq', '')

glasses=glasses.rename(columns={'Comment':'Sample Name','Total':'Total_wt%','Mg Number':'Mg#'})
glasses['Sample_original']=glasses['Sample Name'].astype(str).str.split('_').str[2]

parts = glasses['Sample_original'].astype(str).str.split(')', n=1, expand=True)
glasses['Sample_original'] = parts[1].fillna(parts[0])

parts = glasses['Sample_original'].astype(str).str.split('W', n=1, expand=True)
glasses['Sample_original'] = parts[1].fillna(parts[0])
glasses.insert(0, "Sample_original", glasses.pop("Sample_original"))
glasses_averaged = average_by_col(glasses.copy(),group_col='Sample_original',non_numeric_behavior='skip')
glasses_averaged.insert(0, "Sample_original", glasses_averaged.pop("Sample_original"))

glasses_averaged['Sample'] = glasses_averaged['Sample_original'].astype(str).map({
    '0903': 'UW3',
    '0904': 'UW4',
    '0905': 'UW5',
    '0907': 'UW7',
})
glasses_averaged.insert(0, "Sample", glasses_averaged.pop("Sample"))
glasses_averaged['Reference']="[Wieser, unpublished @ Cambridge Probe lab]"
glasses_averaged
```

### Next, glass data from Sides et al. 2014

```python
Sides_glasses_og=pd.read_excel(MasterFolder+slash+'Data'+slash+'Data_Kea_Uwe_Litt'+slash+'Data_all_EPMA'+
                          slash+'41561_2014_BFngeo2140_MOESM158_ESM.xlsx', sheet_name="Matrix Glass").dropna(axis=0)
Sides_glasses_og['Sample_original']=Sides_glasses_og['Comment'].astype(str).str.split('-g|-og|_g').str[0]

### Fix non numeric columns
Sides_glasses=Sides_glasses_og.apply(pd.to_numeric, errors="coerce")
Sides_glasses[["Sample_original", "Comment", "Date", "Code", "Fountain"]] = Sides_glasses_og[["Sample_original", "Comment", "Date", "Code", "Fountain"]]
Sides_glasses=Sides_glasses.rename(columns={'Comment':'Sample Name','Total':'Total_wt%','Mg Number':'Mg#'})

### Now average
Sides_glasses_averaged = average_by_col(Sides_glasses.copy(),group_col='Sample_original',non_numeric_behavior='skip')
Sides_glasses_averaged.insert(0, "Sample_original", Sides_glasses_averaged.pop("Sample_original"))

Sides_glasses_averaged['Sample'] = Sides_glasses_averaged['Sample_original'].astype(str).map({
    '1700': 'KL0921',
    '1920': 'KL0930',
    '1974-J2': 'KL0924'
})
Sides_glasses_averaged.insert(0, "Sample", Sides_glasses_averaged.pop("Sample"))
## sub just the ones we want
Sides_glasses_averaged_sub=Sides_glasses_averaged[Sides_glasses_averaged['Sample_original'].str.contains("1700|1920|1974-J2")]

Sides_glasses_averaged_sub['Reference']="[Sides et al. 2014]"
display(Sides_glasses)
display(Sides_glasses_averaged_sub)
```

### Now combine the two glass sets
- We also drop the error columns because these are now means, so we report STD instead

```python
combo_glasses=pd.concat([glasses_averaged,Sides_glasses_averaged_sub],axis=0)
combo_glasses=combo_glasses.drop(columns=combo_glasses.columns[combo_glasses.columns.str.contains("_err", case=False)])

combo_glasses.to_excel(compilation_folder+'/'+"UW_L84_K21_averaged.xlsx")
combo_glasses
```
